### **Univariate Analysis**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('/content/diabetes_model.csv')

plt.figure()
sns.histplot(x='Age', data=df,bins=10,color="pink")
plt.title("Histogram-Age")
plt.show()
````

Insight:The histogram provided shows the distribution of age within a sample population. most of the people belongs to the age group of 20-23(youths).the old ages people are very less in this sample set of data.

````python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Glucose', data=df,color='lightgreen')
plt.title('Box Plot-Glucose')
plt.show()
````

From this graph it is clear that  indicating that 50% of the population has a glucose level between 100-150.

## **Bivariate Analysis**

````python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BloodPressure',y='Insulin', data=df,color='red')
plt.title('Scatter Diagram')
plt.show()
````

The scatter plot displays the relationship between Blood Pressure and Insulin levels. There appears to be a concentration of data points with blood pressure around 60-80 and insulin levels below 200, but no strong linear correlation is apparent, with outliers showing higher insulin levels scattered across various blood pressure values.

````python
corr_matrix = df[['Glucose','Insulin']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix,annot=True,cmap='viridis')
plt.title('Correlation Matrix')
plt.show()
````

 It shows a weak positive correlation (0.32) between glucose and insulin levels. Both variables have a perfect correlation with themselves, indicated by a value of 1 on the diagonal.








### **Multivariate Analysis**

````python
sns.pairplot(df[['BloodPressure','Insulin','BMI','Glucose']],hue='BMI',palette='viridis')
plt.title('Pairplot')
plt.show()
````

The pair plot visualizes relationships between Blood Pressure, Insulin, and Glucose, with BMI as the hue, using scatter plots and kernel density estimates along the diagonals. Blood Pressure, Insulin, and Glucose all show clustering of data around certain ranges, particularly with Glucose around 100-150 and Insulin below 200. The scatter plots indicate that there are no strong linear relationships between the variables, with BMI appearing more dispersed across different Glucose levels. The density plots reveal that most data points are concentrated near specific values, particularly low insulin levels, moderate glucose levels, and average blood pressure readings.
